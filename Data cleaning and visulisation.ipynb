Step.1: Import Liberies

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

Step.2 : Create dataset

```python


# Create Dataset
data = {
    "Name": [
        "Amit", "Priya", "Rahul", "Sneha", "Rohan",
        "Anjali", "Vikash", "Neha", "Suman", "Karan",
        "Pooja", "Arjun", "Meena", "Rakesh", "Nisha",
        "Deepak", "Kavita", "Ajay", "Simran", "Ravi"
    ],
    
    "Age": [
        25, 30, np.nan, 28, 35,
        22, 29, np.nan, 31, 27,
        26, 34, 24, 32, np.nan,
        29, 23, 36, 28, 30
    ],
    
    "Gender": [
        "Male", "Female", "Male", "Female", "Male",
        "Female", "Male", "Female", "Female", "Male",
        "Female", "Male", "Female", "Male", "Female",
        "Male", "Female", "Male", "Female", "Male"
    ],
    
    "Salary": [
        40000, 50000, 45000, np.nan, 60000,
        38000, 52000, 48000, np.nan, 47000,
        43000, 62000, 39000, 58000, 45000,
        np.nan, 41000, 65000, 49000, 53000
    ],
    
    "Department": [
        "HR", "IT", "Sales", "Finance", "IT",
        "HR", "Sales", "Finance", "IT", "HR",
        "Sales", "IT", "Finance", "HR", "Sales",
        "IT", "Finance", "HR", "Sales", "IT"
    ],
    
    "Experience": [
        2, 5, 3, 4, np.nan,
        1, 6, 2, 5, 3,
        np.nan, 7, 2, 6, 4,
        5, 1, 8, np.nan, 4
    ]
}

# Convert into DataFrame
df = pd.DataFrame(data)
# Add Duplicate Rows
df = pd.concat([df, df.iloc[[2, 7]]], ignore_index=True)
# Display Dataset
print(df)
# Shape of Dataset
print("\nShape of Da taset:", df.shape)
```

# 5.Missing value visualtion 

```python

plt.figure(figsize=(8,5))
sns.heatmap(df.isnull(), cbar=True, cmap="viridis")

plt.title("Missing Value Visualization")
plt.show()
```

# Step.6 Handle missing values

```python
df["Age"] = df["Age"].fillna(df["Age"].mean())
df["Salary"] = df["Salary"].fillna(df["Salary"].mean())
df["Experience"] = df["Experience"].fillna(df["Experience"].mean())
print(df.isnull().sum())
```

# Step.7: Check duplicate Rows

```python
df.duplicated().sum()
print(df[df.duplicated()])
```

# Step.8: Remove duplicate Rows

```python
df = df.drop_duplicates()
df.duplicated().sum()
```

# Step.9 : Detect Outlirs Using Boxplot

```python
plt.figure(figsize=(6,4))
sns.boxplot(data=df)
plt.title("Boxplot")
plt.show()
```

```python
# Separate Features and Target Variable

X = df.drop("Salary", axis=1)   
y = df["Salary"]             

print("Features (X):")
print(X.head())

print("\nTarget (y):")
print(y.head())
```

```python
# Label Encoding

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df["Gender"] = le.fit_transform(df["Gender"])
df["Department"] = le.fit_transform(df["Department"])

print(df.head())
```

```python
X = df.drop(["Name", "Salary"], axis=1)
y = df["Salary"]

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Train-Test Split

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y,
    test_size=0.2,
    random_state=42
)

print("X_train Shape:", X_train.shape)
print("X_test Shape:", X_test.shape)
print("y_train Shape:", y_train.shape)
print("y_test Shape:", y_test.shape)
```

```python
# Create Age Group

df["Age_Group"] = pd.cut(
    df["Age"],
    bins=[0,18,25,40,130],
    labels=["Child","Young","Adult","Senior"]
)

print(df[["Age","Age_Group"]].head())
```

```python
# Create Performance Column

df["Performance"] = pd.cut(
    df["Salary"],
    bins=[0,40000,50000,60000,70000],
    labels=["Low","Average","Good","Excellent"]
)

print(df[["Salary","Performance"]].head())
```
